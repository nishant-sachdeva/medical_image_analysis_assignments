```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

from skimage import color, io
```

```python
# Question 2

def reinhard_transform(source_image, target_image):
    #conver to lab space
    source = np.copy(source_image)
    target = np.copy(target_image)
    
    lab_source = cv2.cvtColor(source, cv2.COLOR_RGB2LAB)
    lab_target = cv2.cvtColor(target, cv2.COLOR_RGB2LAB)
    
    # now apply the transformation
    l_source = lab_source[:,:,0]
    a_source = lab_source[:,:,1]
    b_source = lab_source[:,:,2]
    
    l_target = lab_target[:,:,0]
    a_target = lab_target[:,:,1]
    b_target = lab_target[:,:,2]
    
    #mean subtract now
    l_source = l_source -  np.mean(l_source)
    a_source = a_source - np.mean(a_source)
    b_source = b_source - np.mean(b_source)
    
    l_target = l_target - np.mean(l_target)
    a_target = a_target - np.mean(a_target)
    b_target = b_target - np.mean(b_target)
    
    # now take ratios of standard deviation
    l_source = (float(np.std(l_target))/np.std(l_source) ) * l_source 
    a_source = (float(np.std(a_target))/np.std(a_source) ) * a_source
    b_source = (float(np.std(b_target))/np.std(b_source) ) * b_source
    
    lab_source[:,:,0] = np.copy(l_source)
    lab_source[:,:,1] = np.copy(a_source)
    lab_source[:,:,2] = np.copy(b_source)
    # now we convert it back to rgb
    return cv2.cvtColor(lab_source, cv2.COLOR_LAB2RGB)


# read source image
source = cv2.imread("../data/color_transfer/source_skin1.jpg")
source = cv2.cvtColor(source, cv2.COLOR_BGR2RGB)


plt.imshow(source)
plt.show()
```

```python

target1 = cv2.imread("../data/color_transfer/target/ISIC_0000030_downsampled.jpg")
target1 = cv2.cvtColor(target1, cv2.COLOR_BGR2RGB)

plt.imshow(target1)
plt.title("Target1")
plt.show()

# plt.imshow(source)
# plt.show()

plt.imshow(reinhard_transform(source, target1))
plt.title("Transformed to target 1")
plt.show()
```
